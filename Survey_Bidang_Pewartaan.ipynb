# Chart BIA/BIR

```python
# Tanggapan anak mengenai kegiatan di BIA/BIR

# PBQ09
# Q: Untuk pertanyaan di bawah ini harus didiskusikan dengan anak Bapak / Ibu. 
#    Bagaimana tanggapan anak Bapak / Ibu mengenai kegiatan di BIA / BIR di bawah ini?"

import seaborn as sns
import matplotlib.pyplot as plt

label_activities = [
    'Kegiatan menyanyi', 
    'Membaca cerita', 
    'Panggung boneka', 
    'Mewarnai', 
    'Permainan / Games', 
    'Mendengarkan Kitab Suci', 
    'Ada snack yang dibagikan', 
    'Gunting Tempel', 
    'Menonton video Kitab Suci'
]

activities = [
    'kegiatan-menyanyi', 
    'kegiatan-cerita', 
    'kegiatan-boneka', 
    'kegiatan-mewarnai',
    'kegiatan-games', 
    'kegiatan-ks', 
    'kegiatan-snack', 
    'kegiatan-tempel', 
    'kegiatan-video'
]

likes = []
dislikes = []
total = []

for activity in activities:
    likes.append(data[data[activity] == 1][activity].count())
    dislikes.append(data[data[activity] == 2][activity].count())
    total.append(likes[-1] + dislikes[-1])

sns.set(rc={"figure.figsize":(10, 6)})
sns.set_palette("deep")
colors = ['#4285F4', '#DB4437', '#F4B400', '#0F9D58']
sns.set_palette(sns.color_palette(colors))
fig, ax = plt.subplots()

ax.barh(label_activities, likes, color='#4285F4', label='Suka')
ax.barh(label_activities, [-x for x in dislikes], color='#DB4437', label='Tidak Suka')

for i, (l, d, t) in enumerate(zip(likes, dislikes, total)):
    ax.annotate(f'{l / t:.0%}', xy=(l / 2, i), va='center', ha='center', color='white')
    ax.annotate(f'{d / t:.0%}', xy=(-d / 2, i), va='center', ha='center', color='white')

ax.set(xlabel="Count", ylabel="Activity", title="Tanggapan Anak mengenai kegiatan di BIA/BIR")
sns.despine(left=True, bottom=True)
plt.legend()
plt.show()
```
